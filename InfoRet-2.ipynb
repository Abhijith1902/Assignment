```python
file = open(r"C:\Users\Dell\Desktop\Stars.txt", encoding='utf8')
file1 = file.read()
file1=file1.split(" ")
file1=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file1])

file = open(r"C:\Users\Dell\Desktop\Galaxy.txt", encoding='utf8')
file2 = file.read()
file2=file2.split(" ")
file2=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file2])

file = open(r"C:\Users\Dell\Desktop\Black Holes.txt", encoding='utf8')
file3 = file.read()
file3=file3.split(" ")
file3=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file3])
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
  
word_tokens = word_tokenize(file1+" "+file2+" "+file3)
  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
 
filtered_sentence=" ".join(filtered_sentence) 
```

```python
lst=[set(filtered_sentence.split(" "))]
unique_words=set().union(*lst)
```

```python
from collections import defaultdict
docs=defaultdict(list)
for word in unique_words:
    if word!=" " or word!='':
        if word in file1.split(" "):
            docs[word].append('1')
        else :
            pass
        if word in file2.split(" "):
            docs[word].append('2')
        else:
            pass
        if word in file3.split(" "):
            docs[word].append('3')
        else:
            pass
    else:
        continue
```

```python
map_dict={'1':'doc1','2':'doc2','3':'doc3'}
```

```python
def query_invert(qry):
    for i in qry.split():
        if i=='and':
            qry= qry.replace(i,'&')
        elif i=='or':
            qry=qry.replace(i,'|')
        elif i=='not':
            qry=qry.replace(i,"{'1','2','3'}-")
    for i in qry.split():
        if i.isalpha():
   
              qry=qry.replace(i,repr(set(docs[i])))
    return qry
```

```python
query="supernova and not blackholes"
```

```python
exp=query_invert(query)
final_val=eval(exp)
final_val
```

```python
for i in final_val:
    print(map_dict[i])
```
